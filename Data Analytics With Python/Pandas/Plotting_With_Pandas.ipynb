## Plotting With Pandas

```python
import math
import collections
import urllib


import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
#sets the maximum number of rows for the data output in pandas
pd.options.display.max_rows = 15
```

```python
gapminder = pd.read_csv('gapminder.csv')
gapminder
```

```python
#Adding an extra column to the Pandas DataFrame dataset for gapminder
gapminder['log_gdp_per_day'] = np.log10(gapminder['gdp_per_capita']/365.25)
```

```python
gapminder.info()
```

```python
gapminder_by_year = gapminder.set_index('year').sort_index()
gapminder_by_year
```

```python
gapminder_by_country = gapminder.set_index('country').sort_index()
gapminder_by_country
```

```python
gapminder_by_year.loc[2005].plot.scatter('log_gdp_per_day','life_expectancy')
pp.show()
```

```python
axes = gapminder_by_year.loc[1960].plot.scatter('log_gdp_per_day','life_expectancy',label = 1960)
gapminder_by_year.loc[2015].plot.scatter('log_gdp_per_day','life_expectancy',label= 2015, color = 'C1', ax = axes)
pp.show()
```

```python
gdp_gapminder = gapminder[['life_expectancy','log_gdp_per_day']]
gdp_gapminder.plot.scatter('life_expectancy','log_gdp_per_day')
pp.show()
```

```python
gapminder_by_country.loc['Italy'].plot('year','life_expectancy')
pp.show()
```

```python
gapminder_by_country.loc['Italy'].sort_values('year').plot('year','life_expectancy')
pp.show()
```

```python
#Comparing three different countries
axes = gapminder_by_country.loc['Italy'].sort_values('year').plot('year','life_expectancy', label='Italy')
gapminder_by_country.loc['China'].sort_values('year').plot('year','life_expectancy',label='China',ax = axes)
gapminder_by_country.loc['United States'].sort_values('year').plot('year','life_expectancy',label='USA',ax = axes)

pp.axis(xmin = 1900)
pp.ylabel('Life Expectancy')
pp.xlabel('Years')

pp.show()
```

```python
#What happened to China in the year 1960 - Disastrous famine
```

```python
#To determine the fertility rate and Life expectancy up to the age of 5 (Globally)
gapminder.babies_per_woman.mean()
```

```python
gapminder.groupby('year').babies_per_woman.mean()
#Women are having fewer babies nowadays compared to the past.
```

```python
gapminder.groupby('year').babies_per_woman.mean().plot()
pp.ylabel('Babies per Woman')
pp.show()
```

```python
gapminder.groupby('year').age5_surviving.mean().plot(secondary_y = True)
pp.ylabel('Age 5 survival [%]')
pp.show()
#There is a rising trend in the survival rate. This could be attributed to 
# the advancement in medical care as well as better nourishment plans for nursing
# mothers with improved GDP_per_capita
```

```python
gapminder.groupby('year').babies_per_woman.mean().plot()
pp.ylabel('Babies per Woman')

gapminder.groupby('year').age5_surviving.mean().plot(secondary_y = True)
pp.ylabel('Age 5 survival [%]')

pp.show()
```

```python
gapminder.pivot_table('babies_per_woman','year','region')
```

```python
gapminder.pivot_table('babies_per_woman','year','region').plot()
pp.show()
#Africa leads in number of babies per woman while Europe is tailing
```

```python
gapminder.pivot_table('age5_surviving','year','region')
```

```python
gapminder.pivot_table('age5_surviving','year','region').plot()
pp.show()

#The trend shows that in Europe there is higher life expectancy of babies surviving up to the age of 5 years
# while the figures in Africa are the lowest. There is a link to the GDP and economic prosperity of the two regions
```
