## Interpolation

```python
xdata = np.array([0,1,4,5,7,8],'d')
ydata = np.array([10,5,2,7,7.5,10],'d')
```

```python
pp.figure(figsize=(10,5))
pp.plot(xdata,ydata,'--o',markersize = 10)
pp.show()
```

```python
xnew = np.linspace(0,8,9)
ynew = np.interp(xnew,xdata,ydata)
```

```python
pp.figure(figsize=(10,5))
pp.plot(xdata,ydata,'--*',ms =10)
pp.plot(xnew,ynew,'--o', ms = 10)
pp.xlabel('X values')
pp.ylabel('Y Values')
pp.grid()
pp.show()
```

```python
#increasing the number of elements in the linspace improves the interpolation
xnew1 = np.linspace(0,8,30)
ynew1 = np.interp(xnew1, xdata,ydata)
```

```python
pp.figure(figsize=(10,5))
pp.plot(xdata,ydata,'--+',ms =10)
pp.plot(xnew,ynew,'--*', ms = 11)
pp.plot(xnew1,ynew1,'--o',ms = 14)
pp.xlabel('X values')
pp.ylabel('Y Values')
pp.grid()
pp.show()
```

#### Using the interpolation knowledge with the PASADENA Missing values dataset


```python
#Loading the broken data afresh
Pasadena = getweather.getyear('PASADENA',['TMIN','TMAX'], 2001)
```

```python
good = ~ np.isnan(Pasadena['TMIN'])  #data points that are not nan
x = np.arange(0,365)  # choose values for the 365 days
np.interp(x,x[good],Pasadena['TMIN'][good])
```

```python
len(Pasadena['TMIN'][good])
```

```python
len(x[good])
```

```python
len(x)
```

```python
#defining a function to automatically fill in the missing values for the 
#weather stations dataset in future applications
def fillnans(array):
    good = ~np.isnan(array)
    x = np.arange(len(array))
    
    return np.interp(x, x[good],array[good])
```

```python
pp.plot(fillnans(Pasadena['TMIN']))
pp.plot(fillnans(Pasadena['TMAX']))
pp.show()
```
